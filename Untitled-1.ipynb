```python
from sklearn.preprocessing import OneHotEncoder

df = pd.get_dummies(df, drop_first=True)  
```

```python
y = df["Price"]  
X = df.drop(columns=["Price"], axis=1) 
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

MODELS AND PREDITCTION

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)
y_pred_lr = lr_model.predict(X_test)
```

```python
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)
y_pred_rf = rf_model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
def evaluate_model(name, y_true, y_pred):
    r2 = r2_score(y_true, y_pred)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    mae = mean_absolute_error(y_true, y_pred)
    
    print(f"\n {name} Model Evaluation")
    print(f"R² Score: {r2:.4f}")
    print(f"RMSE: {rmse:.2f}")
    print(f"MAE: {mae:.2f}")
    
    return r2, rmse, mae
```

```python
r2_lr, rmse_lr, mae_lr = evaluate_model("Linear Regression", y_test, y_pred_lr)
r2_rf, rmse_rf, mae_rf = evaluate_model("Random Forest", y_test, y_pred_rf)
```

```python
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.scatterplot(x=y_test, y=y_pred_lr, alpha=0.6, color="blue")
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Linear Regression: Actual vs Predicted")
```

```python
plt.subplot(1, 2, 2)
sns.scatterplot(x=y_test, y=y_pred_rf, alpha=0.6, color="green")
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Random Forest: Actual vs Predicted")

plt.tight_layout()
plt.show()
```

```python
error_rf = y_test - y_pred_rf
plt.figure(figsize=(8, 6))
sns.histplot(error_rf, bins=30, kde=True, color="red")
plt.xlabel("Prediction Error")
plt.title("Random Forest Prediction Errors Distribution")
plt.show()
```

```python
bins = np.linspace(y.min(), y.max(), 6)  # Create 5 price ranges
y_test_binned = np.digitize(y_test, bins)
y_pred_binned = np.digitize(y_pred_rf, bins)

conf_matrix = confusion_matrix(y_test_binned, y_pred_binned)

plt.figure(figsize=(6,5))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues", xticklabels=bins, yticklabels=bins)
plt.xlabel("Predicted Price Range")
plt.ylabel("Actual Price Range")
plt.title("Confusion Matrix for Price Prediction")
plt.show()
```
